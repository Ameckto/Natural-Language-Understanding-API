```python
import pandas as pd
import requests
import json
import numpy as np
```

```python
def genereate_test_data(input_text_arr):
    data_arr = list()
    for intent_arr in input_text_arr:
        intent_original = intent_arr.pop(0)
        for input_text in intent_arr:
            payload = {'text':input_text}
            headers = {'content-type': 'application/json'}
            response = requests.post('http://localhost:5000/model/parse', json=payload, headers=headers).content
            my_json = response.decode('utf8').replace("'", '"')
            data = json.loads(my_json)
            intent_name = data['intent']['name']
            intent_confidence = data['intent']['confidence']
            entities = data['entities']
            intent_entities = set()
            for entity in entities: 
                entity_value = entity['value']
                intent_entities.add(entity_value)
            data_arr.append([intent_original, input_text, intent_name, intent_confidence, intent_entities])
    return data_arr
```

```python
wake_up = ['wake_up', 'hey bbb', 'hey big blue button', 'whats up my friend bbb', 'hello bbb', 'gey big blue botton']
mute = ['mute', 'mute Steffen', 'mute Seid', 'please mute Klaus', 'can you please mute Mozzam?', 'get Freddy muted please']
summarize = ['summarize', 'make a summary', 'summarize the meeting', 'make me a summarization please', 'get me the summary of the meeting now', 'can I have the summary please?']
give_presenter = ['give_presenter', 'give Mike presenter', 'give me presenter', 'let myself be presenter', 'let me present', 'make Sonya present now please']
share_screen = ['share_screen', 'share my screen please', 'please share my screen', 'can you share my screen now?', 'is there a way to share my screen?', 'can you share nmy screen please?']
raise_hand= ['raise_hand', 'raise my hand', 'raise my hand up please', 'please raise my hand', 'now, raise my hand', 'raise my hands now']

wake_up_give_presenter = ['wake_up+give_presenter']
wake_up_give_presenter.extend([i +' ' + j for i, j in zip(wake_up[1:], give_presenter[1:])])

wake_up_share_screen = ['wake_up+share_screen']
wake_up_share_screen.extend([i +' ' + j for i, j in zip(wake_up[1:], share_screen[1:])])

wake_up_raise_hand = ['wake_up+raise_hand']
wake_up_raise_hand.extend([i +' ' + j for i, j in zip(wake_up[1:], raise_hand[1:])])

wake_up_mute = ['wake_up+mute']
wake_up_mute.extend([i +' ' + j for i, j in zip(wake_up[1:], mute[1:])])

wake_up_summarize =['wake_up+summarize']
wake_up_summarize.extend([i +' ' + j for i, j in zip(wake_up[1:], summarize[1:])])

input_text_arr = list()
input_text_arr.extend([wake_up, mute, summarize, give_presenter, share_screen,
                       raise_hand, wake_up_give_presenter, wake_up_share_screen,
                       wake_up_raise_hand, wake_up_summarize, wake_up_mute
                      ])
```

```python
data = genereate_test_data(input_text_arr)
```

```python
df = pd.DataFrame(data)
df = df.rename(columns={0: 'intent_expected', 1: 'input_text', 2: 'intent_identified', 3: 'intent_confidence', 4: 'intent_entities'})
df['right_intent'] = np.where(df['intent_identified'] == df['intent_expected'], 'true', 'false')
df = df.reindex(columns=['input_text', 'intent_identified', 'intent_expected', 'intent_confidence', 'intent_entities', 'right_intent'])
df
```

```python
df.groupby('intent_expected').right_intent.value_counts().unstack().plot(kind='bar', stacked=True)
```

```python
def genereate_test_data_2(input_arr):
    data_arr = list()
    for intent_arr in input_arr:
        for intent_dict in intent_arr:
            intent_entitiy_string = np.nan
            payload = {'text': intent_dict['input_text']}
            headers = {'content-type': 'application/json'}
            response = requests.post('http://localhost:5000/model/parse', json=payload, headers=headers).content
            my_json = response.decode('utf8').replace("'", '"')
            data = json.loads(my_json)
            intent_name = data['intent']['name']
            intent_confidence = data['intent']['confidence']
            entities = data['entities']
            intent_entities = set()
            for entity in entities: 
                entity_value = entity['value']
                intent_entities.add(entity_value)
            intent_entities = list(intent_entities)
            if len(intent_entities) > 0:
                intent_entitiy_string = intent_entities[0]
                if len(intent_entities) > 1:
                    for string in intent_entities[1:]:
                        intent_entitiy_string += ', ' + string
        
            single_dict = {'input_text': intent_dict['input_text'],
                              'intent_expected': intent_dict['intent_expected'],
                              'intent_identified': intent_name,
                              'intent_confidence': intent_confidence,
                              'entities': intent_entitiy_string   
            }
            
            try: 
                single_dict.update({'entities_expected': intent_dict['entities_expected']})
            except:
                None
            data_arr.append(single_dict)
            
            
                
                
    return data_arr
```
